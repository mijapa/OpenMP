```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
%matplotlib inline
```

```python
times = pd.read_csv('./cmake-build-debug/times.csv',index_col=0 ,header = 0, encoding='latin-1')
times.head(3)
```

```python
g = sns.lineplot(marker="o", dashes=True, data=times)
g.set_title("Times")
g.set(ylabel='Time [s]')
# plt.xticks(times.index)
g.figure.set_size_inches(20,8)
```

```python
sequential = pd.read_csv('./cmake-build-debug/sequential.csv',header = 0, encoding='latin-1')
sequential['sequential'][0]
```

```python
speedup = times
speedup['static'] = sequential['sequential'][0]/speedup['static']
speedup['dynamic'] = sequential['sequential'][0]/speedup['dynamic']
speedup['guided'] = sequential['sequential'][0]/speedup['guided']
speedup.head(3)
```

```python
g = sns.lineplot(marker="o", dashes=True, data=times)
g.set_title("Speedup")
g.set(ylabel='Speedup')
# plt.xticks(times.index)
g.figure.set_size_inches(20,8)
```
